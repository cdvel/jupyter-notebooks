### Understand data with visualizations

```python
from pandas import read_csv
from matplotlib import pyplot
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(url, names=names)
```

#### 1. Univariate plots

##### 1.1. Univariate histograms

a quick way to get an idea of the distribution of each attribute (e.g, gaussian, skewed, exponential) and identify outliers

- age, pedi, test possibly exponential distribution
- mass, plas, press, gaussian or nearly gaussian distribution

**Many ML techniques assume gaussian univariate distributions on the input variables**

```python
data.hist()
pyplot.tight_layout()
pyplot.show()
```

```python
#### 1.2 Univariate density plots

Another representation of attribute distribution 
```

```python
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False, legend=True, fontsize=5)
pyplot.tight_layout()
pyplot.subplots_adjust(top=1.5, right=1.5)
pyplot.show()
```

#### Box and Whisker plots

Summary of distribution (median, 25th and 75th percentile) using boxes, and whiskers help represent the spread of data outside for outliers 

Here

age, skin, test are skewed towards smaller values

```python
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
pyplot.tight_layout()
pyplot.subplots_adjust(top=1.5, right=1.5)
pyplot.show()
```

### Multivariate plots

Represent interaccions between variables

#### Correlation matrix

Relation between variable changes. 
Positive correlation: same direction changes
Negative correlation: oposite directions

A correlation matrix, gives you a sense of which variables are correlated with each other. Linear and logistic regression (between others) perform poorly with highly correlated variables.





```python
import numpy

correlations = data.corr()
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
# Adding attribute names and no. of ticks
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
# end add
pyplot.show()
```

#### Scatter plot

Relationships dot representation in two dimensions. Scatter plot matrix, show relationships for **all** attribute pairs. Useful to spot structured relationships between data (e.g., linear), which may need to be removed.

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(data)
pyplot.subplots_adjust(top=1.5, right=1.5)
pyplot.show()
```
